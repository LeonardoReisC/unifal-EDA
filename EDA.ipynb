# Introdução
No âmbito deste projeto de análise exploratória de dados, exploraremos um conjunto de dados que oferece insights valiosos sobre a satisfação dos passageiros em companhias aéreas. A satisfação do cliente é um aspecto crucial para qualquer empresa, e no setor de aviação, onde as experiências dos passageiros desempenham um papel significativo, compreender os fatores que influenciam a satisfação é essencial. Este conjunto de dados fornece uma visão detalhada dos fatores que impactam a satisfação dos passageiros, permitindo-nos identificar quais elementos específicos desempenham um papel fundamental na construção de uma experiência positiva ou negativa.

dataset-cover.png

A pesquisa de satisfação foi realizada para coletar informações sobre diversos aspectos da experiência do passageiro em voos de companhias aéreas. O objetivo principal é determinar quais fatores estão altamente correlacionados com a satisfação do passageiro, bem como se é possível prever antecipadamente a satisfação do passageiro com base em diversas variáveis. Essas informações são de grande importância para as companhias aéreas, pois podem ajudar a orientar melhorias em serviços e tomar decisões estratégicas para melhor atender às expectativas dos passageiros.

## Atributos:

O conjunto de dados consiste em várias variáveis relevantes que capturam diferentes aspectos da experiência do passageiro. Alguns dos principais atributos incluem:

1. **Gênero:** O sexo dos passageiros, que nos permitirá avaliar se a satisfação varia entre os gêneros.

2. **Tipo de cliente:** Categoriza os passageiros como clientes fiéis ou desleais, o que pode influenciar a satisfação de acordo com a lealdade à companhia aérea.

3. **Idade:** A idade real dos passageiros, que nos permite investigar a influência da idade na satisfação.

4.  **Tipo de Viagem:** Registra a finalidade da viagem (pessoal ou de negócios), o que pode desempenhar um papel importante na satisfação do passageiro.

5. **Classe:** Refere-se à classe de viagem (Business, Eco, Eco Plus) e sua relação com a satisfação.

6. **Distância de voo:** A distância da viagem, que pode afetar a satisfação do passageiro devido à duração da viagem.

7. **Serviço Wi-Fi a bordo:** Avalia o nível de satisfação dos passageiros em relação ao serviço de Wi-Fi disponível durante o voo, variando de 0 (Não Aplicável) a 5 (Muito Satisfeito).  

8. **Hora de partida/chegada conveniente:** Indica o nível de satisfação dos passageiros com os horários de partida e chegada dos voos.  

9.  **Facilidade de reserva online:**  Refere-se à satisfação dos passageiros com o processo de reserva de voos online.

10. **Localização do portão:** Avalia o nível de satisfação dos passageiros em relação à localização dos portões de embarque nos aeroportos.  

11. **Comida e bebida:** Reflete a satisfação dos passageiros em relação à qualidade e variedade de alimentos e bebidas disponíveis durante o voo.  

12. **Embarque online:** Avalia a satisfação dos passageiros com a facilidade e eficiência do processo de embarque online.  

13. **Conforto do assento:** Indica o nível de satisfação dos passageiros em relação ao conforto dos assentos durante o voo.

14. **Entretenimento a bordo:** Refere-se à satisfação dos passageiros em relação às opções de entretenimento disponíveis durante o voo, como filmes, música e jogos.  

15. **Serviço de bordo:** Avalia o nível de satisfação dos passageiros em relação ao serviço prestado pela tripulação durante o voo.  

16. **Conforto para as pernas:** Indica a satisfação dos passageiros com o espaço para as pernas disponível nos assentos.  

17. **Manuseio de bagagem:** Reflete a satisfação dos passageiros em relação ao processo de manuseio de bagagem, incluindo a entrega e cuidado com as malas.  

18. **Serviço de check-in:** Avalia a satisfação dos passageiros em relação ao processo de check-in nos aeroportos.  

19. **Serviço de bordo:** Representa outra avaliação da satisfação dos passageiros em relação ao serviço prestado pela tripulação durante o voo.  

20. **Limpeza:** Indica o nível de satisfação dos passageiros em relação à limpeza geral da aeronave.  

21. **Atraso de partida em minutos:** Registra o tempo de atraso na partida do voo em minutos.  

22. **Atraso de chegada em minutos:** Registra o tempo de atraso na chegada do voo em minutos.  

23. **Satisfação:** É a variável de resposta que indica o nível geral de satisfação dos passageiros com a companhia aérea, classificando-os em duas saídas com base nas avaliações anteriores:
 * Satisfeito;
 * Neutro ou Insatisfeito.

### **Distribuição da Quantidade de Homens e Mulheres**

---


O histograma da quantidade de homens e mulheres oferece uma visualização da distribuição de pessoas com base em seu gênero.

```python
contagem_genero = df['Gender'].value_counts()
plt.figure(figsize=(11, 8))
ax = sns.barplot(x=contagem_genero.index, y=contagem_genero.values, palette=cores_pastel, edgecolor='black')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=12)
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.title('Quantidade de Homens e Mulheres')
plt.show()
```

Observa-se, inicialmente, uma semelhança notável na quantidade de homens e mulheres, com 65.899 mulheres e 63.981 homens. Apesar de uma discreta predominância no número de mulheres, a distribuição dos dados relacionados ao gênero apresenta um equilíbrio considerável.

Neste gráfico de pizza, exploraremos a distribuição proporcional das pessoas com base em seu gênero.

```python
plt.figure(figsize=(7, 7))
plt.pie(contagem_genero, labels=contagem_genero.index, autopct='%1.1f%%', colors=cores_pastel, startangle=90,  wedgeprops={"edgecolor": "black",
                'linewidth': .8,
                'antialiased': True})
plt.title('Porcentagem de Homens e Mulheres')
plt.show()
```

Ao verificar a distribuição proporcional das pessoas com base em seu gênero, observa-se que as mulheres representam aproximadamente 50,7% do total, enquanto os homens constituem 49,3%. Concluímos, portanto, que há um equilíbrio significativo entre homens e mulheres nessa amostra.

### **Nível de Satisfação por Gênero**

---


O Histograma de Satisfação por Gênero ilustra a distribuição da satisfação de passageiros com base em seu gênero. Ele destaca quantos passageiros estão satisfeitos e quantos estão neutros ou não estão satisfeitos com o serviço, distinguindo entre gênero masculino e feminino.

```python
plt.figure(figsize=(15, 7))
ax = sns.histplot(df, x='Gender', hue='satisfaction', multiple="dodge",
                  palette=cores_customizadas, shrink=.8)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title('Histograma do Nível de Satisfação por Gênero')
plt.show()
```

A partir da análise desse gráfico, podemos observar que a maioria das pessoas do sexo feminino, aproximadamente 37.630, expressam insatisfação ou neutralidade com o serviço da companhia, em contraste com cerca de 28.269 que demonstram satisfação. No que diz respeito ao sexo masculino, a maioria também relata neutralidade ou insatisfação com o serviço, enquanto 28.159 indicam satisfação. Assim, independentemente do sexo, o nível de insatisfação ou neutralidade permanece mais alto entre os clientes.

O gráfico de pizza permite uma visualização mais clara da distribuição da satisfação por gênero.

```python
hist_sat_gen = df.groupby('satisfaction')['Gender'].value_counts()
plt.figure(figsize=(15, 5))
plt.pie(
    hist_sat_gen,
    labels=hist_sat_gen.index,
    autopct='%1.1f%%',
    colors=cores_pastel,
    wedgeprops={"edgecolor": "black",
                'linewidth': .8,
                'antialiased': True},
)
plt.axis('equal')
plt.title('Porcentagem do Nível de Satisfação separada por Gênero')
plt.show()
```

O nível de satisfação entre homens e mulheres é semelhante, com ambos em torno de 22%, o que equivale a menos de um quarto do total de passageiros. A parcela de pessoas que se encontram em estado neutro ou insatisfeito representa a maior parte, correspondendo a aproximadamente 28% para homens e 29% para mulheres. ### **Label Encoding**

---


> Obtenção das features categóricas para o Label Encoding


```python
df.info()
```

```python
df.head()
```

Para garantir o correto funcionamento dos modelos que serão utilizados, é essencial realizar a conversão das variáveis categóricas em valores numéricos. Isso permite obter um modelo preciso e resultados apropriados para as análises a serem conduzidas.

> Gerando _encoded data_


```python
encoded_df = df.copy()
non_numerical = ['Gender', 'Customer Type',
                 'Type of Travel', 'Class', 'satisfaction']

l_encoder = LabelEncoder()

for feature in non_numerical:
    encoded_df[feature] = l_encoder.fit_transform(encoded_df[feature])
```

```python
encoded_df.head()
```

### **Análise da distribuição das avaliações mais correlacionadas à Satisfação**

---


Nesta fase, iremos analisar as variáveis relacionadas à experiência do usuário, que são avaliadas em uma escala de 0 a 5 e têm impacto na satisfação dos passageiros.

```python
reviews = ['Inflight wifi service',
           'Departure/Arrival time convenient', 'Ease of Online booking',
           'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
           'Inflight entertainment', 'On-board service', 'Leg room service',
           'Baggage handling', 'Checkin service', 'Inflight service',
           'Cleanliness', 'satisfaction']

most_correlated = encoded_df[reviews].corr(
)['satisfaction'].sort_values(ascending=False)[1:5]
most_correlated
```

Observa-se que as variáveis que apresentam as maiores correlações com a satisfação dos passageiros são:

1. **Check-in online** (Online boarding): 0,501749.
2. **Entretenimento a bordo** (Inflight entertainment): 0,398234.

# **Utilizando o algoritmo KNN nos conjuntos de treino e teste sem aplicar transformações nos dados**

---


Na primeira etapa da classificação do modelo, foi adotado o algoritmo KNN (K-Nearest Neighbors) com um valor de k igual a 10, sem a aplicação de quaisquer transformações nos dados do conjunto de dados. A divisão dos dados consistiu em alocar 75% das instâncias para o conjunto de treinamento, enquanto os 25% restantes foram reservados para o conjunto de teste. Além disso, a métrica escolhida para avaliação do modelo foi a acurácia, devido à sua interpretação direta e fácil compreensão. A acurácia mede a proporção de instâncias corretamente classificadas em relação ao total, fornecendo uma visão geral do desempenho do modelo.

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

def train_and_test(data):
    columns = data.columns
    columns = columns.drop('satisfaction')
    x = data[columns]
    y = data['satisfaction']

    x_train, x_test, y_train, y_test = train_test_split(
        x, y, random_state=0, train_size=.75, stratify=y)

    KNN = KNeighborsClassifier(n_neighbors=10, metric='euclidean')

    KNN.fit(x_train, y_train)

    print("KNN training score :", round(KNN.score(x_train, y_train) * 100, 2))
    print("KNN testing score :", round(KNN.score(x_test, y_test) * 100, 2))
```

> Obtenção do desempenho do modelo


```python
train_and_test(encoded_df)
```

O resultado obtido nessa etapa foi um score de 79,32% para o conjunto de treino e 74,88% para o conjunto de teste. Esses valores são considerados relativamente bons, no entanto, existem oportunidades de melhoria por meio das transformações a serem aplicadas nos dados.

Para corrigir a assimetria na distribuição das distâncias percorridas por voo, optamos por aplicar o PowerTransformer com o método "box-cox". Essa abordagem visa melhorar a simetria dos dados, proporcionando uma representação mais equilibrada das distâncias. A escolha dessa transformação baseia-se na capacidade do método "box-cox" de ajustar a variabilidade dos dados, buscando uma distribuição mais próxima da normal. Este processo visa aprimorar a qualidade da análise estatística das distâncias percorridas pelos voos.

```python
from sklearn.preprocessing import PowerTransformer

flight_scaler_box = PowerTransformer(
    method='box-cox', standardize=True).fit_transform(pd.DataFrame(encoded_df[feature]))
plt.figure(figsize=(15, 5))
ax = sns.histplot(flight_scaler_box, kde=True,
                  kde_kws={'bw_adjust': 5, 'cut': 2}, color=cores_pastel[4])
plt.title('Distribuição das Distâncias Percorridas Por Voo por Box-Cox Scaling')
plt.show()
```
